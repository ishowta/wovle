```python
res = []
for i in range(1,5):
    res.append(pd.read_pickle('../data/processed/csj/'+str(i)+'/grid_result.pd'))
```

```python
for i in range(len(res)):
    res[i] = res[i][['point']]
    maxp = res[i]['point'][0]
    minp = res[i]['point'][-2]
    rangep = maxp - minp
    print(minp)
    res[i] = res[i].apply(lambda x: (x['point']-minp)/rangep, axis=1)
```

```python
p = pd.DataFrame()
```

```python
for (i,r) in enumerate(res):
    p[i] = r
```
